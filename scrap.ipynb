```python
# Import libraries
from bs4 import BeautifulSoup
import requests
import re
import cohere

# Bible story URLs to scrape
story_urls = [
  "https://www.biblegateway.com/passage/?search=John+2&version=NIV", # Water to wine
  "https://www.biblegateway.com/passage/?search=matthew+14%3A13-21&version=NIV", # Feeding 5000
  "https://www.biblegateway.com/passage/?search=luke+15%3A11-32&version=NIV", # Prodigal son
]

# Extract story text from page
def scrape_story(url):
  page = requests.get(url)
  soup = BeautifulSoup(page.content, 'html.parser')
  text = soup.find('div', class_='passage-text').get_text()
  
  # Clean text
  text = re.sub(r"\[(.*?)\]", "", text) # Remove [ ]
  text = re.sub("\s+", " ", text) # Normalize whitespace
  
  return text

# Scrape all stories  
stories = {}
for url in story_urls:
  name = url.split("=")[1] # Get story name from URL
  stories[name] = scrape_story(url) 

# Print stories
for key, story in stories.items():
  print(key)
  print(story)
  print()

# Setup Cohere
co = cohere.Client('q81O6CJT17c7qfSQc7U693xvFwAVf9rDhsjoHv2b')

# Query function
def query_cohere(story_text, question):
  response = co.generate(
    prompt=f"Story: {story_text}\nQuestion: {question}",
    max_tokens=500,
  )

  return response.generations[0].text 

# Sample query
key = "John+2&version"
query = "What is the moral of this story?"
ai_response = query_cohere(stories[key], query)
print(ai_response)

# Log responses
story_logs = {}

for key, story in stories.items():
  query = "Where did it happen? What is the moral of this story?"
  response = query_cohere(story, query)
  
  story_logs[key] = {
    "story": story, 
    "questions": [query],
    "responses": [response]
  }
  
print(story_logs)
```
